```python
import matplotlib.pyplot as plt

# Data preparation (replace with your actual data)
experience_level = [1, 2, 3, 4, 5]  # Example experience levels
salary = [50000, 60000, 70000, 80000, 90000]  # Example corresponding salaries
employee_names = ['John', 'Emily', 'David', 'Sophia', 'Michael']  # Example employee names

# Scatter plot for Experience Level vs. Salary
plt.scatter(experience_level, salary, s=100, c='blue', alpha=0.7)
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.title('Experience Level vs. Salary')

# Annotating data points with employee names
for i in range(len(experience_level)):
    plt.annotate(employee_names[i], (experience_level[i], salary[i]),
                 textcoords="offset points", xytext=(0,10), ha='center')

plt.grid(True)  # Add grid lines
plt.xticks(experience_level)  # Set x-axis tick labels
plt.yticks(range(40000, 100001, 10000))  # Set y-axis tick labels
plt.tight_layout()  # Adjust plot layout

plt.show()
```
